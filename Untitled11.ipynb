## Paso 2: Preprocesamiento de los datos

En esta secci칩n vamos a limpiar los tweets antes de pasarlos a los modelos.  
El objetivo es reducir ruido y normalizar el texto para mejorar el rendimiento en la clasificaci칩n.

### Pasos de preprocesamiento aplicados:
1. **Eliminar URLs**: los links no aportan valor sem치ntico.
2. **Eliminar menciones (`@user`)**: irrelevantes para la emoci칩n.
3. **Eliminar hashtags (`#`)**: mantenemos la palabra, pero quitamos el s칤mbolo.

5. **Convertir a min칰sculas**: normalizaci칩n del texto.
6. **Eliminar espacios extra**.
7. **(Stop Words)**: quitar stopwords en espa침ol.

## Modelo 2: Red Neuronal LSTM con Embeddings

En este segundo modelo utilic칠 una red neuronal recurrente (LSTM) para la clasificaci칩n de emociones.  
A diferencia de la Regresi칩n Log칤stica con TF-IDF, este modelo aprende representaciones distribuidas de las palabras y captura dependencias secuenciales en los tweets.

**Pasos principales:**
1. Tokenizaci칩n de los tweets y conversi칩n a secuencias num칠ricas.
2. Padding de las secuencias para que tengan la misma longitud.
3. Creaci칩n de un embedding layer (con vocabulario aprendido en el dataset).
4. Construcci칩n de un modelo secuencial con:
   - Capa Embedding
   - Capa LSTM
   - Capa densa de salida con activaci칩n softmax (8 clases).
5. Entrenamiento con los conjuntos de entrenamiento y validaci칩n.
6. Evaluaci칩n en los conjuntos de validaci칩n y prueba.
7. Reporte de m칠tricas:
   - Accuracy
   - Precision, Recall y F1-score por emoci칩n
   - Matriz de confusi칩n


```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

max_words = 10000
max_len = 50

tokenizer_lstm = Tokenizer(num_words=max_words)
tokenizer_lstm.fit_on_texts(train_df["clean_tweet"])

X_train_seq = tokenizer_lstm.texts_to_sequences(train_df["clean_tweet"])
X_dev_seq = tokenizer_lstm.texts_to_sequences(dev_df["clean_tweet"])
X_test_seq = tokenizer_lstm.texts_to_sequences(test_df["clean_tweet"])

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_dev_pad = pad_sequences(X_dev_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

y_train_array = np.array(y_train)
y_dev_array = np.array(y_dev)
y_test_array = np.array(y_test)

model_lstm = Sequential()
model_lstm.add(Embedding(max_words, 128, input_length=max_len))
model_lstm.add(LSTM(64, dropout=0.2, recurrent_dropout=0.2))
model_lstm.add(Dense(len(encoder.classes_), activation="softmax"))

model_lstm.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model_lstm.fit(X_train_pad, y_train_array, validation_data=(X_dev_pad, y_dev_array), epochs=3, batch_size=32)

y_pred_lstm = model_lstm.predict(X_test_pad).argmax(axis=1)
print(classification_report(y_test_array, y_pred_lstm, target_names=encoder.classes_))
print(confusion_matrix(y_test_array, y_pred_lstm))
```

## Modelo 3: DistilBERT Multiling칲e (Fine-Tuning)

En este tercer modelo apliqu칠 fine-tuning de **DistilBERT Multiling칲e**, una versi칩n ligera de BERT entrenada en m칰ltiples idiomas, incluido el espa침ol.  
El objetivo fue aprovechar el conocimiento preentrenado del modelo para mejorar la clasificaci칩n de emociones sin necesidad de entrenar desde cero.

Pasos principales:
- Tokenizaci칩n de los tweets usando `DistilBertTokenizer`.
- Conversi칩n a tensores y construcci칩n de datasets para entrenamiento y evaluaci칩n.
- Entrenamiento de un modelo `DistilBertForSequenceClassification` con 8 clases (una por emoci칩n).
- Evaluaci칩n en los conjuntos de validaci칩n y prueba.
- Reporte de m칠tricas de **accuracy, precisi칩n, recall, F1-score** por emoci칩n, adem치s de la matriz de confusi칩n.



```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import classification_report
import torch
from torch.utils.data import Dataset

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-multilingual-cased")
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-multilingual-cased", num_labels=len(encoder.classes_))

class TweetDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=64):
        self.texts = list(texts)
        self.labels = list(labels)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encodings = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        item = {key: val.squeeze(0) for key, val in encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

train_dataset = TweetDataset(train_df["clean_tweet"], y_train, tokenizer)
dev_dataset = TweetDataset(dev_df["clean_tweet"], y_dev, tokenizer)
test_dataset = TweetDataset(test_df["clean_tweet"], y_test, tokenizer)

training_args = TrainingArguments(
    output_dir="./results_distilbert",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    save_strategy="epoch",
    logging_dir="./logs_distilbert",
    logging_steps=20
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=dev_dataset,
)

trainer.train()

preds = trainer.predict(test_dataset)
y_pred_labels = preds.predictions.argmax(axis=1)

print(classification_report(y_test, y_pred_labels, target_names=encoder.classes_))
```
