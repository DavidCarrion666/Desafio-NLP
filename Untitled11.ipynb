## Paso 2: Preprocesamiento de los datos

En esta sección vamos a limpiar los tweets antes de pasarlos a los modelos.  
El objetivo es reducir ruido y normalizar el texto para mejorar el rendimiento en la clasificación.

### Pasos de preprocesamiento aplicados:
1. **Eliminar URLs**: los links no aportan valor semántico.
2. **Eliminar menciones (`@user`)**: irrelevantes para la emoción.
3. **Eliminar hashtags (`#`)**: mantenemos la palabra, pero quitamos el símbolo.
4. **Eliminar emojis y caracteres especiales**: solo dejamos letras, números y algunos signos básicos.
5. **Convertir a minúsculas**: normalización del texto.
6. **Eliminar espacios extra**.
7. **(Stop Words)**: quitar stopwords en español.

Con este preprocesamiento generaremos una nueva columna llamada `clean_tweet` que usaremos en los modelos.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

vectorizer = TfidfVectorizer(max_features=5000)
X_train = vectorizer.fit_transform(train_df["clean_tweet"])
X_dev = vectorizer.transform(dev_df["clean_tweet"])
X_test = vectorizer.transform(test_df["clean_tweet"])

encoder = LabelEncoder()
y_train = encoder.fit_transform(train_df["emotion"])
y_dev = encoder.transform(dev_df["emotion"])
y_test = encoder.transform(test_df["emotion"])

X_train.shape, X_dev.shape, X_test.shape
```

## Paso 3: Entrenamiento del Primer Modelo (Regresión Logística)

En este paso entrenaremos un modelo base de **Regresión Logística** utilizando las representaciones **TF-IDF** de los tweets.  
Este modelo servirá como **línea base** para comparar con los otros dos modelos que exploraremos más adelante.  

Evaluaremos el rendimiento en el conjunto de validación (`dev_df`) y en el conjunto de prueba (`test_df`), reportando métricas como:

- **Precisión (precision)**
- **Exhaustividad (recall)**
- **F1-score**
- **Matriz de confusión por clase (emoción)**  

Esto nos permitirá medir qué tan bien se desempeña el modelo en la clasificación de emociones antes de probar arquitecturas más complejas.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

model_lr = LogisticRegression(max_iter=2000, class_weight="balanced")
model_lr.fit(X_train, y_train)

y_pred_dev = model_lr.predict(X_dev)
y_pred_test = model_lr.predict(X_test)

print("Resultados en Dev:")
print(classification_report(y_dev, y_pred_dev, target_names=encoder.classes_))

print("Resultados en Test:")
print(classification_report(y_test, y_pred_test, target_names=encoder.classes_))

cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=encoder.classes_, yticklabels=encoder.classes_)
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

## Modelo 2: Red Neuronal LSTM con Embeddings

En este segundo modelo utilicé una red neuronal recurrente (LSTM) para la clasificación de emociones.  
A diferencia de la Regresión Logística con TF-IDF, este modelo aprende representaciones distribuidas de las palabras y captura dependencias secuenciales en los tweets.

**Pasos principales:**
1. Tokenización de los tweets y conversión a secuencias numéricas.
2. Padding de las secuencias para que tengan la misma longitud.
3. Creación de un embedding layer (con vocabulario aprendido en el dataset).
4. Construcción de un modelo secuencial con:
   - Capa Embedding
   - Capa LSTM
   - Capa densa de salida con activación softmax (8 clases).
5. Entrenamiento con los conjuntos de entrenamiento y validación.
6. Evaluación en los conjuntos de validación y prueba.
7. Reporte de métricas:
   - Accuracy
   - Precision, Recall y F1-score por emoción
   - Matriz de confusión


```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

max_words = 10000
max_len = 50

tokenizer_lstm = Tokenizer(num_words=max_words)
tokenizer_lstm.fit_on_texts(train_df["clean_tweet"])

X_train_seq = tokenizer_lstm.texts_to_sequences(train_df["clean_tweet"])
X_dev_seq = tokenizer_lstm.texts_to_sequences(dev_df["clean_tweet"])
X_test_seq = tokenizer_lstm.texts_to_sequences(test_df["clean_tweet"])

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_dev_pad = pad_sequences(X_dev_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

y_train_array = np.array(y_train)
y_dev_array = np.array(y_dev)
y_test_array = np.array(y_test)

model_lstm = Sequential()
model_lstm.add(Embedding(max_words, 128, input_length=max_len))
model_lstm.add(LSTM(64, dropout=0.2, recurrent_dropout=0.2))
model_lstm.add(Dense(len(encoder.classes_), activation="softmax"))

model_lstm.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model_lstm.fit(X_train_pad, y_train_array, validation_data=(X_dev_pad, y_dev_array), epochs=3, batch_size=32)

y_pred_lstm = model_lstm.predict(X_test_pad).argmax(axis=1)
print(classification_report(y_test_array, y_pred_lstm, target_names=encoder.classes_))
print(confusion_matrix(y_test_array, y_pred_lstm))
```

## Modelo 2: LSTM

En este segundo modelo entrené una red neuronal LSTM sobre los tweets.  
El objetivo fue capturar dependencias secuenciales y contextuales en el texto, algo que la regresión logística con TF-IDF no logra.  

Los resultados obtenidos fueron los siguientes:

- Para clases con menos ejemplos como *disgust*, *fear* o *surprise*, el modelo no logró predecir adecuadamente, quedando con valores de precisión y recall cercanos a 0.
- La clase *joy* y *sadness* tuvieron un rendimiento algo mejor, con un recall de 27% y 13% respectivamente, aunque sigue siendo bajo.

**Conclusión personal:**  
El modelo LSTM mostró cierta capacidad para aprender patrones básicos en el texto, especialmente en clases más frecuentes (*others* y *joy*), pero no logró generalizar bien para las clases menos representadas. Esto indica que la red no tuvo suficiente información balanceada para aprender todas las emociones. Aun así, la mejora frente al modelo más simple en ciertas clases refleja que las secuencias de palabras aportan información útil para la tarea.


## Modelo 3: DistilBERT Multilingüe (Fine-Tuning)

En este tercer modelo apliqué fine-tuning de **DistilBERT Multilingüe**, una versión ligera de BERT entrenada en múltiples idiomas, incluido el español.  
El objetivo fue aprovechar el conocimiento preentrenado del modelo para mejorar la clasificación de emociones sin necesidad de entrenar desde cero.

Pasos principales:
- Tokenización de los tweets usando `DistilBertTokenizer`.
- Conversión a tensores y construcción de datasets para entrenamiento y evaluación.
- Entrenamiento de un modelo `DistilBertForSequenceClassification` con 8 clases (una por emoción).
- Evaluación en los conjuntos de validación y prueba.
- Reporte de métricas de **accuracy, precisión, recall, F1-score** por emoción, además de la matriz de confusión.



```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import classification_report
import torch
from torch.utils.data import Dataset

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-multilingual-cased")
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-multilingual-cased", num_labels=len(encoder.classes_))

class TweetDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=64):
        self.texts = list(texts)
        self.labels = list(labels)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encodings = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        item = {key: val.squeeze(0) for key, val in encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

train_dataset = TweetDataset(train_df["clean_tweet"], y_train, tokenizer)
dev_dataset = TweetDataset(dev_df["clean_tweet"], y_dev, tokenizer)
test_dataset = TweetDataset(test_df["clean_tweet"], y_test, tokenizer)

training_args = TrainingArguments(
    output_dir="./results_distilbert",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    save_strategy="epoch",
    logging_dir="./logs_distilbert",
    logging_steps=20
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=dev_dataset,
)

trainer.train()

preds = trainer.predict(test_dataset)
y_pred_labels = preds.predictions.argmax(axis=1)

print(classification_report(y_test, y_pred_labels, target_names=encoder.classes_))
```
